# Multiple Linear regression

## Import

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.pylab import rcParams
rcParams["figure.figsize"] = 15, 6
import seaborn as sns
sns.set()
```

## Target function:
\begin{align*}
    f(x) = \frac{\sin\left(\pi x \right)}{\pi x}
\end{align*}

in here, 
$x \in [-3,3] $, and $\varepsilon \sim U(0,1)$    
observation err $0.05 \times \varepsilon$

## Random data generation with target function

```python
def target_func(x):
    return np.sin(np.pi * x) / (np.pi * x)

def data_gen(n=50, N=1000, linspace=False, seed=0):
    if linspace == False:
        np.random.seed(seed)
        x = np.random.uniform(-3, 3, n)
    else:
        x = np.linspace(-3, 3, n)
    X = np.linspace(-3, 3, N)
    y = target_func(x) + 0.05 * np.random.randn(n)
    Y = target_func(X)
    return x, y, X, Y
```

### Plotting

```python
x, y, X, Y = data_gen()
plt.scatter(x, y)
plt.scatter(X, Y)
plt.legend(["sample", "real_function"])
```

## Generate Trigonometric
\begin{align*}
    \Phi = \begin{pmatrix}
    \phi_{1} (x_{1}) & \cdots & \phi_{b}(x_{1}) \\
    \vdots & \ddots & \vdots \\
    \phi_{1} (x_{n}) & \cdots & \phi_{b}(x_{n})
    \end{pmatrix}
    \quad
    \text{or}
    \quad
    \Phi = \begin{pmatrix}
    1 & \phi_{1} (x_{1}) & \cdots & \phi_{b}(x_{1}) \\
    \vdots &　\vdots &　\ddots & \vdots \\
    1 & \phi_{1} (x_{n}) & \cdots & \phi_{b}(x_{n})
    \end{pmatrix}
\end{align*}
here, 
\begin{align*}
    \mathbf{\phi}(x) = \left(\sin\frac{x}{2}, \cos\frac{x}{2}, \sin\frac{2x}{2}, \cos\frac{2x}{2}, \ldots, \sin\frac{15x}{2}, \cos\frac{15x}{2}, \right)^{\top}
\end{align*}

```python
def get_trigonometric(x, X, n_base=30, intercept=True):
    if intercept == True:
        p = np.zeros((len(x), n_base))
        P = np.zeros((len(X), n_base))
        p = np.hstack([np.array([1]*(len(x))).reshape(-1,1), p])
        P = np.hstack([np.array([1]*(len(X))).reshape(-1,1), P])
        for j in np.arange(1, n_base/2 + 1):
            p[:, int(2 * j - 1)] = np.sin(x * j / 2)
            p[:, int(2 * j)] = np.cos(x * j / 2)
            P[:, int(2 * j - 1)] = np.sin(X * j / 2)
            P[:, int(2 * j)] = np.cos(X * j / 2)
    else:
        p = np.zeros((len(x), n_base))
        P = np.zeros((len(X), n_base))
        for j in np.arange(1, n_base / 2 + 1):
            p[:, int(2 * j - 2)] = np.sin(x * j / 2)
            p[:, int(2 * j - 1)] = np.cos(x * j / 2)
            P[:, int(2 * j - 2)] = np.sin(X * j / 2)
            P[:, int(2 * j - 1)] = np.cos(X * j / 2)    
    return p, P
```

## solve the normal equation
\begin{align*}
    \Phi^{\top}\Phi \mathbf{w} = \Phi^{\top} \mathbf{y}
\end{align*}

```python
phi_train, phi_pred = get_trigonometric(x, X, intercept=True)
w = np.linalg.solve(np.dot(phi_train.T, phi_train), np.dot(phi_train.T, y))
y_pred = np.dot(phi_pred, w)
```

```python
plt.ylim(-0.4, 1.1)
plt.scatter(x, y)
plt.plot(X, Y, '--')
plt.plot(X, y_pred)
```

Gaussian kernel matrix

$
\begin{align*}
    　k(x, \mu) = \exp\left(-\frac{\| x - \mu \|^{2}}{2\sigma^2} \right) 
\end{align*}
$

$
\begin{align*}
    K = \begin{pmatrix}
    k(x_{1}, x_{1}) & \cdots & k(x_{1}, x_{n})\\
    \vdots & \ddots & \vdots \\
    k(x_{n}, x_{1}) & \cdots & k(x_{n}, x_{n})
    \end{pmatrix}
\end{align*}
$

\begin{align*}
    K^TK \mathbf{w} = K^T\mathbf{y}
\end{align*}

```python
def gaussian_kernel_matrix(x, X, sigma=0.3):
    x = x.reshape(-1, 1)
    X = X.reshape(-1, 1)
    xx = x ** 2
    XX = X ** 2
    ss = 2 * sigma ** 2
    
    k = np.exp(-(np.tile(xx, (1, len(x))) + np.tile(xx.T, (len(x), 1)) - 2* x * x.T) / ss)
    K = np.exp(-(np.tile(XX, (1, len(x))) + np.tile(xx.T, (len(X), 1)) - 2* X * x.T) / ss)
    return k, K
```

```python
k_train, k_pred = gaussian_kernel_matrix(x, X, sigma=1)
KtK = np.dot(k_train.T, k_train)
KtY = np.dot(k_train.T, y)
w = np.linalg.solve(KtK, KtY)
y_pred = k_pred@w

plt.ylim(-0.4, 1.1)
plt.scatter(x, y)
plt.plot(X, Y, '--')
plt.plot(X, y_pred)
```
